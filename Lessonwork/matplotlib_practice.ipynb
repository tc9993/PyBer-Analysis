```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis , y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis , y_axis, color='green', marker='D', linewidth=2, label="Boston")
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Fare ($)')
ax.set_xlabel('Date')
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Create plot
plt.plot(x_axis , y_axis)
plt.show()
```

```python
# Create the plot and add label for legend
plt.plot(x_axis , y_axis, label = 'Boston', marker='*', color='blue', linewidth=2)

#Create Labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

#Y limit beteween 0 and 45
plt.ylim(0, 45)

# x limit
plt.xlim('Jan' , 'Dec')

#Create title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()

# Add a grid
plt.grid()
```

```python
plt.plot(x_axis , y_axis , 'o')
```

```python
plt.scatter(y_axis , x_axis , color='red' , label='Chicago')
plt.gca().invert_yaxis()
plt.legend()
plt.title('PyBer Fare by Month')
plt.ylabel('Fare ($)')
plt.xlabel('Date')
plt.xlim(0,50)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis , y_axis , s= [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis] , color='skyblue' , alpha=.5 , edgecolor='black' , linewidth=2, label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.invert_yaxis()
ax.set_xlim(0, 50)
```

```python
plt.pie(y_axis , labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values,
        labels=x_axis, 
        autopct='%.1f%%', 
        colors=colors)
plt.show()
```

```python
fig , ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, 
       labels=x_axis, 
       explode=explode_values, 
       colors=colors, 
       autopct='%.1f%%',
       shadow = True,
       startangle = -90)
ax.invert_yaxis()
plt.show()
```
